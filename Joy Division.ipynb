I really like the album artwork of Joy Division's Unknown Pleasures. I want to see if I can turn regular images into something like that. Let's take an image of a butterfly because why not.

```python
#Importing all dependencies

import matplotlib.image as img 
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
from scipy import ndimage
from PIL import Image
```

```python
path = r"C:\Users\APOORV\Downloads\cc26bbeae94298f0045513cd5daa0453.jpg"
picture = Image.open(path)
rad = np.asarray(picture)
rad = rad[:,:,0]
rad = 255*rad
```

```python
sad = []
import math
t = int(pow(math.sqrt(np.shape(rad)[0]), .65))
c = len(rad);
for i in range(len(rad)):
    if i%t==0:
        sad.append(rad[i]/255+c)
        c-=1
r = np.arange(0, len(sad[0]))
ax = plt.axes()
# Setting the background color
ax.set_facecolor("black")

for i in sad:
    plt.plot(r, i, 'white')
```
