```python
!pip install opencv-python
!pip install scikit-image

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
#path = "D:/"
#name = "wow patil wow"

path = "C:/New Drive/Movies/Animated/The Wind Rises/"
name = "The.Wind.Rises.2013.DVDRIP.XviD"
output_path = "D:/P.N.I.A/Illustrator/Inspiration/Movie Viz/"

video = cv2.VideoCapture(path + name + ".avi")
```

```python
hours = 2
mins = 1
secs = 29
rate_frames = 23.98
target_frames = 1000
dim_frames = (352,704) 

time_secs = ((((hours * 60) + mins) * 60) + secs)
vid_frames = rate_frames * time_secs
skipping_every = int(vid_frames / target_frames)


print("Number of frames: " + str(vid_frames))
print("Target number of frames: " + str(target_frames))
print("Skipping every " + str(skipping_every) + " images")
np.ceil(vid_frames/skipping_every)
```

```python
# MEAN

video = cv2.VideoCapture(path + name + ".avi")

count = 0
success = True
image_mean = []
success, image = video.read()

temp = []
for _ in range(3):
    temp.append(np.mean(image[:,:,_]))
image_mean = np.asarray([[temp]]*dim_frames[0])
print(np.shape(image_mean))
count += 1
#lol = np.asarray(lol)


while success:
    #cv2.imwrite("video_data/image_%d.jpg" % count, image)    
    success, image = video.read()
    if count % skipping_every == 0:
        temp = []
        for _ in range(3):
            temp.append(np.mean(image[:,:,_]))
        image_mean = cv2.hconcat([image_mean, np.asarray([[temp]]*dim_frames[0])])
        print(np.shape(image_mean))
    count += 1
```

```python
# Saving the MEAN image

filename_png = output_path + name + "/" + "mean_" + str(target_frames) + "_png" + ".png"
filename_jpg = output_path + name + "/" + "mean_" + str(target_frames) + "_jpg" + ".jpg"

y = cv2.UMat(np.array(image_mean, dtype=np.uint8))

cv2.imwrite(filename_jpg, y)
cv2.imwrite(filename_png, y)
```

```python
# INTERPOLATION

video = cv2.VideoCapture(path + name + ".avi")

count = 0
success = True
#lol = np.asarray(lol)


while success:   
    success, image = video.read()
    if count == 0:
        print(np.shape(image))
        image_resized = resize(image, (dim_frames[0],1), anti_aliasing = False)
        lol = image_resized
    else:
        if count % skipping_every == 0:
            print(np.shape(lol))
            image_resized = resize(image, (dim_frames[0], 1), anti_aliasing = False)
            lol = cv2.hconcat([lol, image_resized])
    count += 1

if np.max(lol) < 2:
    lol = 255 * lol
else:
    pass
```

```python
# Saving the INTERPOLATED image

filename_png = output_path + name + "/" + "interpolated_" + str(target_frames) + "_png" + ".png"
filename_jpg = output_path + name + "/" + "interpolated_" + str(target_frames) + "_jpg" + ".jpg"

x =cv2.UMat(np.array(lol, dtype=np.uint8))

cv2.imwrite(filename_jpg, x)
cv2.imwrite(filename_png, x)
```

```python
# MEDIAN

video = cv2.VideoCapture(path + name + ".avi")

count = 0
success = True
image_median = []
success, image = video.read()

temp = []
for _ in range(3):
    temp.append(np.median(image[:,:,_]))
image_median = np.asarray([[temp]]*dim_frames[0])
print(np.shape(image_median))
count += 1
#lol = np.asarray(lol)


while success:
    #cv2.imwrite("video_data/image_%d.jpg" % count, image)    
    success, image = video.read()
    if count % skipping_every == 0:
        temp = []
        for _ in range(3):
            temp.append(np.median(image[:,:,_]))
        image_median = cv2.hconcat([image_median, np.asarray([[temp]]*dim_frames[0])])
        print(np.shape(image_median))
    count += 1
```

```python
# Saving the MEDIAN image

filename_png = output_path + name + "/" + "median_" + str(target_frames) + "_png" + ".png"
filename_jpg = output_path + name + "/" + "median_" + str(target_frames) + "_jpg" + ".jpg"

z = cv2.UMat(np.array(image_median, dtype=np.uint8))

cv2.imwrite(filename_jpg, z)
cv2.imwrite(filename_png, z)
```
